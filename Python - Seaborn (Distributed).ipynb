# Seaborn :Distributed plot

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df=sns.load_dataset("tips")
```

```python
df.head(3)
```

```python
df.tail(3)
```

```python
sns.relplot( data=df, x="total_bill", y="tip", kind ="scatter", col="sex")
```

```python
sns.relplot(data=df,x="total_bill",y="tip",col="smoker",row="sex")
```

```python
sns.relplot(data=df,x="total_bill",y="tip",col="smoker",row="day")
```
