# KDE Plot:

Kernel Density Estimation (KDE) is a non-parametric method to estimate the probability density function (PDF) of a random variable. Unlike histograms that represent data with bins, KDE smooths the observed data points to create a continuous estimate of the distribution.

Concept:

The KDE is a sum of kernel functions centered at each data point. The kernel functions are smooth, symmetric, and typically unimodal. The most commonly used kernel is the Gaussian kernel, but other kernels such as Epanechnikov, uniform, and triangular can also be used.

```python
sns.kdeplot(data=df,x="total_bill")
```
