# Heatmap:


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
data=px.data.gapminder()
df=sns.load_dataset("tips")
dft=sns.load_dataset("titanic")
```

```python
tdf=data.pivot(index="country",columns="year",values="lifeExp")
sns.heatmap(tdf)
plt.figure(figsize=(15,10))
```

```python
temp_df=data.pivot(index="country",columns="year",values="pop")
```

```python
sns.heatmap(temp_df)
```

```python
temp =data[data["continent"]== "Europe"].pivot(index="country",columns="year",values="lifeExp")
sns.heatmap(temp)
```

```python
temp_asia=data[data["continent"]=="Asia"].pivot(index="country",columns="year",values="lifeExp")
sns.heatmap(temp_asia)
```

```python
sns.heatmap(temp,annot=True)
```

```python
sns.heatmap(temp_asia,annot=True,cmap="viridis")
```

```python
sns.heatmap(temp_asia,annot=True,cmap="summer")
```

```python
sns.heatmap(temp_asia,annot=True,cmap="autumn")
```

```python
sns.heatmap(temp_asia,annot=True,cmap="jet")
```
